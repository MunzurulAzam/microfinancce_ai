```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import os
import warnings
warnings.filterwarnings('ignore')


filename = "/Users/smini/Downloads/flutterWorkStation/ai/pycharm/data/master_data_for_ml.csv"


if os.path.exists(filename):

    df = pd.read_csv(filename)

    print("Data loaded successfully!")
    print(f"Shape: {df.shape}")
    print("\nFirst few rows:")
    print(df.head())
else:
    print(f"Error: '{filename}' ফাইলটি পাওয়া যায়নি। নিশ্চিত করুন এটি আপনার প্রজেক্ট ফোল্ডারে আছে।")
```

```python
# Data Preprocessing

# Data cleaning
def preprocess_data(df):
    # Copy dataframe
    data = df.copy()

    # Handle missing values
    data.fillna('Unknown', inplace=True)

    # Convert date columns
    data['disbursementDate'] = pd.to_datetime(data['disbursementDate'])

    # Create performance metrics
    data['repayment_rate'] = data['totalPayment'] / data['loanAmount']
    data['is_overdue'] = data['OverdueCollectionCount'] > 0
    data['loan_age_days'] = (pd.to_datetime('today') - data['disbursementDate']).dt.days

    # Client performance score
    def calculate_client_score(row):
        score = 100
        score -= row['OverdueCollectionCount'] * 10
        if row['repayment_rate'] >= 1: score += 20
        elif row['repayment_rate'] >= 0.8: score += 10
        score += row.get('cycle', 0) * 5
        return max(0, min(100, score))

    data['client_performance_score'] = data.apply(calculate_client_score, axis=1)

    return data

# Preprocess data
df_processed = preprocess_data(df)
print("Data preprocessing completed!")
print(f"New columns added: {[col for col in df_processed.columns if col not in df.columns]}")
```
